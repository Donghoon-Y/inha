```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

```python
#2-a)
zero = np.array([-0.5])
pole = np.array([0, -10, -50])
gain = 2000
G = zpk(zero, pole, gain)

bode(G)
fig = plt.gcf().set_size_inches(8, 8)
# plt.xlim([-r, r]), plt.ylim([-r, r])
plt.grid()
plt.show()
```

```python
#2-b)
zero = np.array([-1])
pole = np.array([0, -2, -4 + np.sqrt(48)*1j, -4 - np.sqrt(48)*1j])
gain = 1000
G = zpk(zero, pole, gain)
bode(G)
fig = plt.gcf().set_size_inches(8, 8)
plt.grid()
plt.show()
```

```python
#2-c)
# 분자: 4s(s + 10) = 4s^2 + 40s
num1 = [1, 0]        # s
num2 = [1, 10]       # s + 10
num = 4 * np.convolve(num1, num2)  # [1, 0] * [1, 10] = [1, 10, 0]

# 분모: (s + 50)(4s^2 + 5s + 4)
den1 = [1, 50]
den2 = [4, 5, 4]
den = np.convolve(den1, den2)

# 전달함수 정의
G = tf(num, den)

# Bode Plot
bode(G)
fig = plt.gcf().set_size_inches(8, 8)
plt.grid()
plt.show()
```

```python
#2-d)
num = np.array([1,2])
den = np.array([1, 20, 0,0])

G = tf(num, den)

bode(G)
fig = plt.gcf().set_size_inches(8, 8)
plt.grid()
plt.show()
```

```python
#2-e)
num1 = [1, 0.5]
num2 = [1, 1.5]
num = np.convolve(num1, num2) 

s     = [1, 0]
squad = [1, 2, 2]
s5    = [1, 5]
s15   = [1, 15]
den = np.convolve(s, squad)
den = np.convolve(den, s5)
den = np.convolve(den, s15)
G = tf(num, den)
bode(G)

fig = plt.gcf().set_size_inches(8, 8)
plt.grid()
plt.show()
```
